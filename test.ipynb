```python
from polygon import RESTClient

import datetime
import os.path

import numpy as np
import pandas as pd
```

```python
print(os.path.isfile("/Users/roy/.ssh/roy_poly_key"))

with open("/Users/roy/.ssh/roy_poly_key") as k:
    important = k.read().strip()

client = RESTClient(str(important))
```

```python
#Returns all of the tickers supported by the Polygon.io API.
tickers = []
for t in client.list_tickers(
	market="stocks",
	active="true",
	order="asc",
	limit="100",
	sort="ticker",
	):
    tickers.append(t)

#Returns all of the types of tickers supported by the Polygon.io API
types = client.get_ticker_types(
	asset_class="stocks",
	locale="us"
	)

print(tickers[0])
print(types[0])
```

```python
for i in types:
    print(i.code, ":", i.description)
```

```python
for i in tickers:
    print(i.ticker, ":", i.name)

print("Returned", len(tickers), "available tickers using the Polygon.io Stock API.")
```

```python
aggs = []
for a in client.list_aggs(
    "AAPL",
    1,
    "day",
    "2023-01-09",
    "2025-02-10",
    adjusted="true",
    sort="asc",
    limit=120,
):
    aggs.append(a)

print(aggs)
```
